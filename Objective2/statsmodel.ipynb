```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
r = np.array([1000, 2000, 800, 500])
nr = np.array([1500, 500, 300])
print(r)
print(nr)
```

```python
def wei(vect):
    wei_vec = []
    for i in range(len(vect)):
        wei_vec.append(vect[i]*(len(vect)/np.sum(vect)))
    print(wei_vec)
```

```python
wei(r)
wei(nr)
wei_r = np.array(wei(r))
wei_nr = np.array(wei(nr))
```

```python
import numpy as np
from statsmodels.stats.weightstats import ttest_ind

# generate two sets of random data with weights
sample1 = np.random.normal(loc=10, scale=3, size=5)
sample2 = np.random.normal(loc=12, scale=3, size=5)
#weights1 = np.random.uniform(size=1000)
#weights2 = np.random.uniform(size=1000)
wei(sample1)
wei(sample2)
wei_1 = np.array(wei(sample1))
wei_2 = np.array(wei(sample2))
```

```python
# perform two-sample t-test with unequal weights
t_stat, p_val, df = ttest_ind(sample1, sample2, usevar='unequal',alternative='larger', weights=(wei(sample1), wei(sample2)), value=0)

# print results
print("t-statistic:", t_stat)
print("p-value:", p_val)
if p_val <0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis") 
```

```python
import numpy as np
from statsmodels.stats.weightstats import ttest_ind


r = np.array([1000, 2000, 800, 500])
nr = np.array([1500, 500, 300])
wei(r)
wei(nr)

t_stat, p_val, df = ttest_ind(r, nr, usevar='unequal', alternative='two-sided', weights=(wei(r), wei(nr)), value=0)

# print results
print("t-statistic:", t_stat)
print("p-value:", p_val)
if p_val <0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```
